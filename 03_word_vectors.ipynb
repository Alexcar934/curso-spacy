##### Palabras más similares a Country 

O se usan en contextos similares, por lo que hay cercanía dimensional

```python
your_word = "country"

ms = nlp.vocab.vectors.most_similar(
    np.asarray([nlp.vocab.vectors[nlp.vocab.strings[your_word]]]), n=10)
words = [nlp.vocab.strings[w] for w in ms[0][0]]
distances = ms[2]
print(words)
```

```python
doc1 = nlp("I like salty fries and hamburgers.")
doc2 = nlp("Fast food tastes very good.")

print(doc1, "<->", doc2, doc1.similarity(doc2))
```

```python
doc3 = nlp("The Empire State Building is in New York")

print(doc3.similarity(doc1))
```

```python
doc4 = nlp("I enjoy oranges.")
doc5 = nlp("I enjoy apples.")

print(doc4.similarity(doc5))
```

La similitud semántica es una medida del grado en que dos fragmentos de texto (como palabras, frases o documentos) están relacionados en significado o contexto.


```python
doc6 = nlp("I enjoy videogames.")

doc5.similarity(doc6)

#La similaridad baja, por lo que la similaridad semática entre palabras influye
```

```python
# Similarity of tokens and spans
french_fries = doc1[2:4]
burgers = doc1[5]
print(french_fries, "<->", burgers, french_fries.similarity(burgers))
```
