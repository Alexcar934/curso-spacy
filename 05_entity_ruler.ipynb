# Entity Ruler

```python
import spacy

nlp= spacy.load("en_core_web_sm")
text = "West Chestertenfieldville was referenced in Mr. Deeds."

doc = nlp(text)
```

Entity Ruler: 

Two different ways to add in custom features to a spacy language pipeline:
* rules based aprroach: Se usan cuando quieres generar un conjunto de reglas basadas en una lista de cosas conocidas, regex o lingui features. (E.g. dates)
* machine learning based approach: (E.g. names) Necesitarías una lista con todos los nombres del mundo, por ejemplo, y con esto no hace falta.

Un nlp practitioner tiene que saber cuando usar rules based y cuando ml

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
ruler = nlp.add_pipe("entity_ruler")
```

```python
nlp.analyze_pipes()
```

```python
patterns = [
    {"label":"PERSON", "pattern":"West Chestertenfieldville"}
]
```

```python
ruler.add_patterns(patterns)
```

```python
doc2 = nlp(text)
for ent in doc2.ents:
    print(ent.text, ent.label_)

    #No ha cambiado porque si nos fijamos en los pipes del nlp nuestro entity_ruler viene después del NER
```

```python
nlp2 = spacy.load("en_core_web_sm")
```

```python
nlp2.add_pipe("entity_ruler", before="ner")
```

```python
ruler.add_patterns(patterns)
```

```python
doc = nlp2(text)
```

```python
for ent in doc.ents:
    print(ent.text, ent.label_)

    #No ha cambiado porque si nos fijamos en los pipes del nlp nuestro entity_ruler viene después del NER
```

```python
nlp2.analyze_pipes()
```

```python
nlp3 = spacy.load("en_core_web_sm")
```

```python
ruler = nlp3.add_pipe("entity_ruler", before='ner')

patterns = [
    {"label":"LOC", "pattern":"West Chestertenfieldville"},
    {"label":"FILM", "pattern":"Mr. Deeds."}
]
```

```python
ruler.add_patterns(patterns)
```

```python
doc = nlp3(text)
```

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```
