# Named Entity Recognition

Es como un tipo de extracción de datos e info de textos --> NER

Como acceder a estos trozos de info

```python
import spacy
nlp = spacy.load('en_core_web_sm')
with open("data/wiki_us.txt", "r") as f:
    text = f.read()
```

```python
doc = nlp(text)
```

```python
for ent in doc.ents:
    print(ent.text, ent.label_) #label_ GPE Geopolitical Entity, LOC general location, CARDINAL, NORP (National or Religious Political), DATE, 
```

### LOS SMALL MODELS pueden tener fallos
ejemplo: the American Revolutionary War --> ORG debería ser event

```python
from spacy import displacy

displacy.render(doc, style="ent")
```
